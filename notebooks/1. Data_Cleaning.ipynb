# Data Cleaning

## World Bank DataFrame

```python
wb = pd.read_csv("../raw_data/1 Indicadores_del_desarrollo_humano_mundial Banco Mundial.csv", encoding='utf-8')
wb.sample(frac=0.1) # sample 10% of total data
```

```python
wb.shape # show rows, colums
```

```python
wb.dtypes # the numbers should be float type, they will be changed later
```

We can see there are values like ".." that doesn't count like null but they should. They will be replaced later in this notebook for null values so they can be counted and analyzed easily. 

```python
wb.isnull().sum() # notice this null count didn't incluede non-numerical values like ".."
```

### Replacing bad values and analyzing null values

```python
wb = wb.replace(to_replace="..", value=np.nan)
```

```python
(
    wb
    .isnull()
    .transpose()
    .pipe( # pipe function is used to pass the transposed DataFrame to the lambda function
        lambda df: (
        sns.heatmap(
            data=df
            )
        )
    )
)
```

```python
(
    wb
    .isnull()
    .melt()
    .pipe(
        lambda df: (
        sns.displot(
            data=df,
            y = "variable",
            hue = "value",
            multiple = "fill",
            aspect = 1  
            )
        )
    )
)
```

We can notice that most of Column 2019 values are null, so it can be dropped, along with Series Code because it won't be needed.

### Deleting columns that aren't goint to be used

```python
wb = wb.drop(columns=["Series Code", "2019 [YR2019]"]) 
```

### Unpivoting columns named like years 

Having every year as attribute would difficult that analysis of the dataset, so we need to unpivot the column and make one column of Year as attribute.

```python

wb_melted = (wb
             .melt(
                    id_vars=['Series Name', 'Country Name', 'Country Code'], # variables that will remain as they are in the reshaped DataFrame
                    var_name='Year', # column name of the original DataFrame that is being melted
                    value_name='Value' # name of the column that will store the values from the original DataFrame that are being melted.
                    )
              ) 
```

```python
wb_melted
```

### Cleaning Year column

Before changing its type, we need to delete the strings on the side of the years.

```python
wb_melted["Year"] = wb_melted["Year"].str[:4]
wb_melted.head()
```

```python
wb_melted.dtypes
```

### Changing Types 

```python
wb_melted["Value"] = wb_melted["Value"].replace(to_replace=np.nan, value="0") 
```

```python
wb_melted["Value"] = wb_melted["Value"].astype(float)
wb_melted["Year"] = wb_melted["Year"].astype(int)
```

```python
wb_melted.dtypes
```

```python
wb_melted
```

There are a lot null values but we don't need to deal with those because we won't use every series

### Pivoting 'Series Name' rows to columns

Series like Population, GDP, unemployemnt, etc. are stored in the column "Series Name", so we need to do the inverse operation that we did with the years columns.

```python
index_columns = ['Country Name', 'Country Code', 'Year']

wb_pivoted = (
    wb_melted
    .pivot_table(index = index_columns, # index of the new df, the resulting pivoted DataFrame will have a multi-level index
                  columns='Series Name', # each unique value in the 'Series Name' column will become a separate column in the pivoted DataFrame.
                  values='Value' # the content of the new columns
                )

)
```

```python
wb_pivoted
```

### Selecting the columns to be used

The current dataframe has a lot of column, so we need to choose some of them that could provide insights in future analysis. If we need other columns later, we always can re-visit this step.

```python
columns_list = wb_pivoted.columns.tolist()
print(columns_list) 
```

```python
selected_columns = [
                    "Crecimiento del PIB (% anual)", 
                    "Desempleo, total (% de la población activa total) (estimación modelado OIT)",
                    "INB per cápita, PPA (a $ internacionales actuales)",
                    "INB per cápita, método Atlas (US$ a precios actuales)",
                    'INB, PPA (a $ internacionales actuales)', 
                    'INB, método Atlas (US$ a precios actuales)',
                    'Inflación, índice de deflación del PIB (% anual)',
                    'PIB (US$ a precios actuales)',
                    'Población activa, total',   
                    ]
```

```python
wb_processed = wb_pivoted[selected_columns]
```

```python
wb_processed.sample(frac=0.1)
```

```python
wb_processed.dtypes
```

### Reseting Index

The current dataframe have a multi-level index so it will be reseted to a more appropiate format

```python
wb_final = wb_processed.reset_index(level=['Country Code', 'Country Name', "Year"]) # these levels of the index will be reset, and the corresponding columns will be added back as regular columns
wb_final.head()
```

### Exporting the DF

```python
wb_final.to_csv('../processed_data/World Bank Indicators.csv', index=False, encoding='utf-8')
```

## Videogames Sales DataFrame

```python
vg_sales = pd.read_csv("../raw_data/4. Video Games Sales.csv", encoding='utf-8')
vg_sales.sample(frac=0.1)
```

```python
vg_sales.dtypes
```

```python
vg_sales.isnull().sum()
```

```python
vg_sales["NA_Sales"] = vg_sales["NA_Sales"].fillna(0)
vg_sales["EU_Sales"] = vg_sales["EU_Sales"].fillna(0)

vg_sales["Other_Sales"] = vg_sales["Other_Sales"].fillna(0)
vg_sales["Global_Sales"] = vg_sales["Global_Sales"].fillna(0)
```
